# Displaying most recently created image

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
import os

image_directory = "tmp"  # Name of temporary directory to store images in

list_of_files = glob.glob('{0}/*'.format(image_directory)) # * means all if need specific format then *.csv
```

```python
if len(list_of_files)>0:
    print("Right-click (on Macs, control-click) on image to download image.")
    latest_file = max(list_of_files, key=os.path.getctime)
    image = mpimg.imread(latest_file)
    fig = plt.figure(1, facecolor='w', edgecolor='w')
    plt.axis('off')
    fig.set_size_inches(28,20)
    plt.imshow(image)
    plt.show()
else:
    print("Error: no image currently available to display")
```
